# Метод скользящей средней
* Simple Moving Average (SMA) — простая скользящая средняя;

* Weighted Moving Average (WMA) — взвешенная скользящая средняя;

* Exponential Moving Average (EMA) — экспоненциальная скользящая средняя.

$$ 
\bar{y1} = \frac{y1+y2+y3}{3};
\bar{y2} = \frac{y2+y3+y4}{3};
\bar{y3} = \frac{y3+y4+y5}{3};
$$

*** Example ***
* Sample input: data=[1,2,3,4,5,6,7], N=2
* Sample output: [1.5, 2.5, 3.5, 4.5, 5.5, 6.5] 

```python
import numpy as np 
import matplotlib.pyplot as plt 
  
def sliding_average(data, period):
    j = next(i for i, x in enumerate(data) if x is not None)
    _range = range(len(data))[j + period - 1:]
    sub_result = [np.mean(data[i - period + 1: i + 1]) for i in _range]
    return np.array(sub_result)

data = [10,12,3,41,15,6,71]
period = 2
sma = sliding_average(data, period)

plt.plot(data)
plt.plot(sma)
plt.show()
    
```
